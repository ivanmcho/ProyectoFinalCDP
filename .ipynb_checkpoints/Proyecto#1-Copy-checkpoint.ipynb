```python


def entrenar_regresion_lineal(x, y, epochs, imprimir_error_cada, learning_rate):
    # Crear una matriz de diseño con la variable independiente x y una columna de unos
    X = np.column_stack((x, np.ones_like(x)))
    
    # Inicializar los parámetros del modelo (β0 y β1) con valores aleatorios
    parametros = np.random.rand(2)
    
    # Función de costo (MSE)
    def calcular_error(predicciones, y):
        return np.mean((predicciones - y) ** 2)
    
    # Entrenamiento del modelo
    for epoch in range(epochs):
        # Calcular las predicciones del modelo
        predicciones = np.dot(X, parametros)
        
        # Calcular el error
        error = calcular_error(predicciones, y)
        
        # Actualizar los parámetros utilizando el descenso de gradiente
        gradiente = np.dot(X.T, (predicciones - y)) / len(y)
        parametros -= learning_rate * gradiente
        
        # Imprimir el error cada imprimir_error_cada iteraciones
        if (epoch + 1) % imprimir_error_cada == 0:
            print(f"Iteración {epoch + 1}: Error = {error:.4f}")
    
    return parametros

# Ejemplo de uso
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])
epochs = 1000
imprimir_error_cada = 100
learning_rate = 0.01

parametros_entrenados = entrenar_regresion_lineal(x, y, epochs, imprimir_error_cada, learning_rate)
print("Parámetros del modelo entrenado:", parametros_entrenados)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

class ModeloRegresionLineal:
    def __init__(self):
        self.modelo_manual = None
        self.modelo_sklearn = None
        self.x_train = None
        self.y_train = None
    
    def entrenar_modelo_manual(self, x, y, epochs, imprimir_error_cada, learning_rate):
        self.modelo_manual = {}
        self.x_train = x
        self.y_train = y
        
        X = np.column_stack((x, np.ones_like(x)))
        parametros = np.random.rand(2)
        
        for epoch in range(epochs):
            predicciones = np.dot(X, parametros)
            error = np.mean((predicciones - y) ** 2) / 2
            gradiente = np.dot(X.T, (predicciones - y)) / len(y)
            parametros -= learning_rate * gradiente
            
            if (epoch + 1) % imprimir_error_cada == 0:
                self.modelo_manual[epoch] = parametros.copy()
                print(f"Iteración {epoch + 1}: Error = {error:.4f}")
                
    def graficar_modelo_evolutivo(self, historial_modelos, n):
        plt.figure(figsize=(10, 6))
        plt.scatter(self.x_train, self.y_train, color='blue', label='Datos de entrenamiento')
        
        # Iterar sobre las iteraciones del historial de modelos
        for iteracion, parametros in historial_modelos.items():
            if (iteracion + 1) % n == 0:
                y_pred = parametros[0] * self.x_train + parametros[1]
                plt.plot(self.x_train, y_pred, label=f'Iteración {iteracion + 1}')
        
        plt.xlabel('Variable independiente')
        plt.ylabel('Variable dependiente')
        plt.title('Evolución del modelo entrenado')
        plt.legend()
        plt.show()
    
    def entrenar_modelo_sklearn(self, x, y):
        self.modelo_sklearn = LinearRegression().fit(x.reshape(-1, 1), y)
    
    def predecir_combinado(self, x):
        pred_manual = np.dot(np.column_stack((x, np.ones_like(x))), self.modelo_manual[max(self.modelo_manual.keys())])
        pred_sklearn = self.modelo_sklearn.predict(x.reshape(-1, 1))
        pred_combinado = (pred_manual + pred_sklearn) / 2
        return pred_manual, pred_sklearn, pred_combinado
    
    def calcular_error(self, x_val, y_val):
        pred_manual, pred_sklearn, pred_combinado = self.predecir_combinado(x_val)
        error_manual = np.mean((pred_manual - y_val) ** 2)
        error_sklearn = np.mean((pred_sklearn - y_val) ** 2)
        error_combinado = np.mean((pred_combinado - y_val) ** 2)
        return error_manual, error_sklearn, error_combinado
    
    def graficar_error(self, x_val, y_val):
        errores = self.calcular_error(x_val, y_val)
        nombres_modelos = ['Manual', 'Sklearn', 'Combinado']
        
        plt.bar(nombres_modelos, errores)
        plt.xlabel('Modelo')
        plt.ylabel('Error')
        plt.title('Error de cada modelo')
        plt.show()


##

modelo = ModeloRegresionLineal()

# Seleccionar las columnas deseadas del DataFrame
x_train = df['OverallQual'].values  # Usar 'OverallQual' como variable independiente
y_train = df['SalePrice'].values     # Usar 'SalePrice' como variable dependiente

# Entrenar el modelo manualmente
modelo.entrenar_modelo_manual(x_train, y_train, epochs=1000, imprimir_error_cada=100, learning_rate=0.01)

# Entrenar el modelo con scikit-learn
modelo.entrenar_modelo_sklearn(x_train, y_train)

# Obtener el error de cada modelo
x_val = df['OverallQual'].values  # Usar 'OverallQual' para datos de validación
y_val = df['SalePrice'].values    # Valores reales de 'SalePrice' para datos de validación
modelo.graficar_error(x_val, y_val)
```

```python
import numpy as np
import matplotlib.pyplot as plt

class RegresionLineal:
    def __init__(self):
        self.beta = np.zeros(2)  # Vector de parámetros del modelo [beta_1, beta_0]
        self.errores = []  # Vector para almacenar los errores de cada iteración
        self.modelos = {}  # Diccionario para almacenar el modelo resultante
    
    def entrenar(self, x, y, epochs, imprimir_error_cada, alpha):
        # Crear una matriz extendida de X con una columna de unos
        X = np.vstack((x, np.ones_like(x))).T
        
        m = len(x)
        for epoch in range(epochs):
            # Calcular las predicciones con el modelo actual de manera vectorizada
            y_pred = np.dot(X, self.beta)
            
            # Calcular el error cuadrático medio (MSE)
            error = (1 / (2 * m)) * np.sum((y_pred - y) ** 2)
            
            # Almacenar el error de esta iteración
            self.errores.append(error)
            
            # Almacenar el modelo resultante de esta iteración
            self.modelos[epoch] = np.copy(self.beta)
            
            # Calcular el gradiente del error con respecto a cada parámetro
            gradient = (1 / m) * np.dot((y_pred - y), X)
            
            # Actualizar los parámetros del modelo utilizando el descenso de gradiente
            self.beta -= alpha * gradient
            
            if epoch % imprimir_error_cada == 0:
                print(f"Iteración: {epoch}, Error: {error}")
                print(f"Parámetros del modelo: {self.beta}")
        
        return self.modelos, self.errores
    
    def predecir(self, x):
        # Crear una matriz extendida de X con una columna de unos
        X = np.vstack((x, np.ones_like(x))).T
        return np.dot(X, self.beta)
    
    
    def visualizar_error(self):
        iteraciones = range(len(self.errores))
        plt.figure(figsize=(10, 6))
        plt.plot(iteraciones, self.errores, marker='o', linestyle='-')
        plt.title('Evolución del error durante el entrenamiento')
        plt.xlabel('Iteración')
        plt.ylabel('Error')
        plt.grid(True)
        plt.show()
        
    def visualizar_modelo(self, modelos, n, x_entrenamiento, y_entrenamiento):
        iteraciones = sorted(list(modelos.keys()))
        iteraciones_seleccionadas = iteraciones[::n]
        
        plt.figure(figsize=(10, 6))
        plt.scatter(x_entrenamiento, y_entrenamiento, color='blue', label='Datos de entrenamiento')
        
        for iteracion in iteraciones_seleccionadas:
            beta_1, beta_0 = modelos[iteracion]
            y_pred = beta_1 * x_entrenamiento + beta_0
            plt.plot(x_entrenamiento, y_pred, label=f'Iteración {iteracion}')
        
        plt.title('Evolución del modelo durante el entrenamiento')
        plt.xlabel('Variable independiente (x)')
        plt.ylabel('Variable dependiente (y)')
        #plt.legend()
        plt.grid(True)
        plt.show()
    
    def entrenar_modelo_sklearn(self, x, y):
        self.modelo_sklearn = LinearRegression().fit(x.reshape(-1, 1), y)
    
    def predecir_sklearn(self, x):
        return self.modelo_sklearn.predict(x.reshape(-1, 1))
    
    def predecir_combinado(self, x):
        pred_manual = self.predecir(x)
        pred_sklearn = self.predecir_sklearn(x)
        pred_combinado = (pred_manual + pred_sklearn) / 2
        return pred_manual, pred_sklearn, pred_combinado
    
    def calcular_error(self, x_val, y_val):
        pred_manual, pred_sklearn, pred_combinado = self.predecir_combinado(x_val)
        error_manual = np.mean((pred_manual - y_val) ** 2)
        error_sklearn = np.mean((pred_sklearn - y_val) ** 2)
        error_combinado = np.mean((pred_combinado - y_val) ** 2)
        return error_manual, error_sklearn, error_combinado
    
    def graficar_error(self, x_val, y_val):
        errores = self.calcular_error(x_val, y_val)
        nombres_modelos = ['Manual', 'Sklearn', 'Combinado']
        
        plt.bar(nombres_modelos, errores)
        plt.xlabel('Modelo')
        plt.ylabel('Error')
        plt.title('Error de cada modelo')
        plt.show()
        
# Ejemplo de uso
#x = np.array([1, 2, 3, 4, 5])
#y = np.array([2, 3, 4, 5, 6])


# Seleccionar las columnas deseadas del DataFrame
x = df['OverallQual'].values  # Usar 'OverallQual' como variable independiente
x2 = df['1stFLrSF'].values
y = df['SalePrice'].values     # Usar 'SalePrice' como variable dependiente


modelo = RegresionLineal()
modelos_entrenamiento, errores_entrenamiento = modelo.entrenar(x, y, epochs=1000, imprimir_error_cada=100, alpha=0.01)

# Visualizar el error durante el entrenamiento
modelo.visualizar_error()

# Visualizar la evolución del modelo cada 3 iteraciones
modelo.visualizar_modelo(modelos_entrenamiento, n=3, x_entrenamiento=x, y_entrenamiento=y)

## modelo2
modelo2 = RegresionLineal()
modelos_entrenamiento2, errores_entrenamiento2 = modelo2.entrenar(x, y, epochs=1000, imprimir_error_cada=100, alpha=0.01)

# Visualizar el error durante el entrenamiento
modelo2.visualizar_error()

# Visualizar la evolución del modelo cada 3 iteraciones
modelo2.visualizar_modelo(modelos_entrenamiento, n=3, x_entrenamiento=x2, y_entrenamiento=y)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

class RegresionLineal:
    def __init__(self):
        self.beta = np.zeros(2)  # Vector de parámetros del modelo [beta_1, beta_0]
    
    def entrenar_modelo_sklearn(self, x, y):
        self.modelo_sklearn = LinearRegression().fit(x.reshape(-1, 1), y)
    
    def predecir_combinado(self, x):
        pred_manual = np.dot(np.column_stack((x, np.ones_like(x))), self.beta)
        pred_sklearn = self.modelo_sklearn.predict(x.reshape(-1, 1))
        pred_combinado = (pred_manual + pred_sklearn) / 2
        return pred_manual, pred_sklearn, pred_combinado
    
    def calcular_error(self, x_val, y_val):
        pred_manual, pred_sklearn, pred_combinado = self.predecir_combinado(x_val)
        error_manual = np.mean((pred_manual - y_val) ** 2)
        error_sklearn = np.mean((pred_sklearn - y_val) ** 2)
        error_combinado = np.mean((pred_combinado - y_val) ** 2)
        return error_manual, error_sklearn, error_combinado

    def graficar_error(self, errores):
        nombres_modelos = ['Manual', 'Sklearn', 'Combinado']
        
        plt.bar(nombres_modelos, errores)
        plt.xlabel('Modelo')
        plt.ylabel('Error')
        plt.title('Error de cada modelo')
        plt.show()

# Seleccionar las variables deseadas del DataFrame
variables_independientes = ['OverallQual', '1stFLrSF']
variable_dependiente = 'SalePrice'

# Verificar si las columnas están presentes en el DataFrame
if all(variable in df.columns for variable in variables_independientes) and variable_dependiente in df.columns:
    # Dividir el conjunto de datos en entrenamiento y validación
    x = df[variables_independientes].values
    y = df[variable_dependiente].values
    x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)

    # Entrenar modelo de regresión lineal de scikit-learn para cada variable
    modelos_sklearn = {}
    for variable in variables_independientes:
        modelo = RegresionLineal()
        modelo.entrenar_modelo_sklearn(x_train[:, variables_independientes.index(variable)], y_train)
        modelos_sklearn[variable] = modelo

    # Calcular errores de cada modelo para las variables seleccionadas
    errores_modelos = {}
    for variable in variables_independientes:
        modelo = modelos_sklearn[variable]
        error_manual, error_sklearn, error_combinado = modelo.calcular_error(x_val[:, variables_independientes.index(variable)], y_val)
        errores_modelos[variable] = {'Manual': error_manual, 'Sklearn': error_sklearn, 'Combinado': error_combinado}

    # Graficar el error de cada modelo para cada variable x
    plt.figure(figsize=(10, 6))
    for variable, errores in errores_modelos.items():
        plt.bar([f'{variable} - {modelo}' for modelo in errores.keys()], list(errores.values()), label=variable)

    plt.xlabel('Variable independiente - Modelo')
    plt.ylabel('Error')
    plt.title('Errores de cada modelo para cada variable x')
    plt.xticks(rotation=45)  # Rotar los labels del eje x
    plt.legend()
    plt.show()
else:
    print("Una o más columnas especificadas no están presentes en el DataFrame.")
```
